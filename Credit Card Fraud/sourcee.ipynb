```python
legit= credit_card_data[credit_card_data.is_fraud == 0]
fraud= credit_card_data[credit_card_data.is_fraud == 1]
```

```python
legit_sample = legit.sample(n=7506)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['is_fraud'].value_counts()
```

```python
new_dataset=new_dataset.drop(columns=['trans_num','dob','job','state','merchant','first','last','street','trans_date_trans_time','city'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
new_dataset['gender']=le.fit_transform(new_dataset['gender'])
new_dataset['category']=le.fit_transform(new_dataset['category'])
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rs=RandomForestClassifier(n_estimators=50)
```

```python
X=new_dataset.drop('is_fraud',axis=1)
Y=new_dataset['is_fraud']
```

```python
rs.fit(X,Y)
```
